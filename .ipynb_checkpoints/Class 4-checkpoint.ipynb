```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None
        
    def append(self, item):
        if not self.head:
            self.head = Node(item)
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = Node(item)
            
    def insert(self, index, value):
        new_node = Node(value)
        if not self.head:
            self.head = new_node
        else:
            if index == 0:
                temp_node = self.head
                self.head = new_node
                new_node.next = temp_node
            else:
                counter = 0
                prev_node = None
                current = self.head
                while current.next and index != counter:
                    counter += 1
                    prev_node = current
                    current = current.next
                if not current.next:
                    current.next = new_node
                else:
                    prev_node.next = new_node
                    new_node.next = current
                    
    def remove(self, value):
        current = self.head
        if current.data == value:
            self.head = current.next
        else:
            prev_node = None
            while current.next and current.data != value:
                prev_node = current
                current = current.next
            if current.data == value:
                prev_node.next = current.next
            else:
                raise ValueError("%s not in list" % repr(value))
                
    def __len__(self):
        counter = 0
        current = self.head
        while current:
            counter += 1
            current = current.next
        return counter
    
    def __getitem__(self, index):
        counter = 0
        current = self.head
        while current and index != counter:
            counter += 1
            current = current.next
        if not current:
            raise IndexError("list index out of range")
        return current.data
            
    def count(self, index):
        current = self.head
        count = 0
        while(current is not None):
            if current.data == index:
                count += 1
            current = current.next
        return count
            
    def __str__(self):
        out_str = "["
        current = self.head
        while current.next:
            out_str += "%s, "  % repr(current.data)
            current = current.next
        out_str += "%s]" % current.data
        return out_str
    
    
```

```python
class DNode:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None
        
    def append(self, value):
        new_node = DNode(value)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
            new_node.prev = current
            
    def insert(self, index, value):
        new_node = DNode(value)
        if not self.head:
            self.head = new_node
        else:
            if index == 0:
                temp_node = self.head
                self.head = new_node
                temp_node.prev = self.head
                new_node.next = temp_node
            else:
                counter = 0
                prev_node = None
                current = self.head
                next_node = current.next
                while next_node and index != counter:
                    counter += 1
                    prev_node = current
                    current = next_node
                if not next_node:
                    current.next = new_node
                else:
                    prev_node.next = new_node
                    new_node.next = current
                    
    def __str__(self):
        out_str = "["
        current = self.head
        while current.next:
            out_str += "%s, "  % repr(current.data)
            current = current.next
        out_str += "%s]" % current.data
        return out_str

mylist = ["a", "b", "c", "d"]                    
dll = DoublyLinkedList()

for i in range(5):
    dll.append(i)

dll.insert(2,"new")
    
print(dll)
```

```python
# Hash Collisions

#  classical hash function

x = 196

modulus = x % 16

print(modulus)

values = [0 for i in range(16)]
keys = [0 for i in range(16)]


def fib(n):
    if n < 2:
        return n
    return fib(n-1) + fib(n-2)
N = 8
values[N%16] = fib(N)

print(values)

modulus_of_n = N%16

if values[modulus_of_n]:
    print("The fibonacci number is: %s" % values[modulus_of_n])
else:
    print("The fibonacci number is: %s" % fib(N))
    
```
